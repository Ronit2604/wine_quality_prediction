**Define Target Variable (y) and Feature Variables (X)**

```python
y=data['quality']
x=data[['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar',
       'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density',
       'pH', 'sulphates', 'alcohol']]
```

**Train Test Split**

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=2529)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

**Modeling**

```python
#SVM
from sklearn.svm import SVC
svc=SVC()
svc.fit(x_train,y_train)
```

```python
#random forest
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier()
model.fit(x_train,y_train)
```

```python
y_pred=svc.predict(x_test)#svm
```

```python
y_pred#svm
```

```python
z_pred=model.predict(x_test)#random forest
z_pred
```

**Model Evaluation**

```python
from sklearn.metrics import accuracy_score #svm
accuracy_score(y_test,y_pred)
```

```python
from sklearn.metrics import accuracy_score #random forest
accuracy_score(y_test,z_pred)
```
